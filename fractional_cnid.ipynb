```python
def calc_frantional(cnids):
    cnids["invcnidy1"] = cnids["cnidv2z"]/(cnids["cnidv1y"]*cnids["cnidv2z"]-cnids["cnidv1z"]*cnids["cnidv2y"])
    cnids["invcnidz1"] = -cnids["cnidv2y"]/(cnids["cnidv1y"]*cnids["cnidv2z"]-cnids["cnidv1z"]*cnids["cnidv2y"])
    cnids["invcnidy2"] = -cnids["cnidv1z"]/(cnids["cnidv1y"]*cnids["cnidv2z"]-cnids["cnidv1z"]*cnids["cnidv2y"])
    cnids["invcnidz2"] = cnids["cnidv1y"]/(cnids["cnidv1y"]*cnids["cnidv2z"]-cnids["cnidv1z"]*cnids["cnidv2y"])
    cnids["cnid1"] = cnids["invcnidy1"]*cnids["dy"] + cnids["invcnidz1"]*cnids["dz"]
    cnids["cnid2"] = cnids["invcnidy2"]*cnids["dy"] + cnids["invcnidz2"]*cnids["dz"]
```

```python
calc_frantional(df_Cu_100)
```
